```python
# Given a string containing digits from 2-9 inclusive,
# return all possible letter combinations that the number could represent.
# Return the answer in any order.
# A mapping of digit to letters (just like on the telephone buttons) is given below.
# Note that 1 does not map to any letters.
```

```python
# storing the letters corresponding to the numbers on the dial screen; notice 7 and 9 contains 4 letters each
import itertools

dic = {}
let = ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
cnt = 0
for i in range(2,10):
    if i==7:
        dic[i]= let[cnt:cnt+4]
        cnt += 4
    elif i==9:
        dic[i]= let[cnt:cnt+4]
        cnt += 4
    else:
        dic[i]= let[cnt:cnt+3]
        cnt += 3
    
    
# taking the input values from keyboard with some exception handling
new_let = [[''],]
nums = input('how many numbers do you wanna input:')
for i in range(int(nums)):
    for j in range(2):
        get = input('\nenter the numbers between 2 and 9 for whom you want the combination of letters:')
        if int(get) == 1 or int(get) == 0 :
            print('invalid input')
        else:
            store = dic[int(get)]
            new_let.append(store)
            break
            


# calling the corresponding letters set and finding all the possible combinations
if len(new_let)>1:
    for l in range(10):
        for i in range(1):
            emp = []
            for j in new_let[i]:
                for k in new_let[i+1]:
                    emp.append(j+k)
            new_let.pop(i)
        new_let[i] = emp
        if len(new_let) == 1:
            break
else:
    print("you have exhausted your number of trials :/")

    
print('\n',new_let,'\n______here is our letter combinations of phone numbers_____ :o')
```
